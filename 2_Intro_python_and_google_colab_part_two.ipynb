### DataFrame

A DataFrame is a high-level data structure composed of indexed series. In practice, it is a table similar to the ones we are used to in Excel, organized into rows and columns.


```python
import numpy as np
import pandas as pd

n = 100
rows = 10
columns = 5

df = pd.DataFrame(np.random.randint(0, n, size=(rows, columns)), columns=[f'var_{i}' for i in range(columns)])
df.insert(loc=0, column='patient_id', value=[f'patient_id_{id}' for id in range(0, rows)])
df.head(10)
```

#### Logical operators

* AND &
* OR |

```python
logical_df = pd.DataFrame()
logical_df['a'] = [True, True, False, False]
logical_df['b'] = [True, False, True, False]
logical_df
```

```python
logical_df['AND'] = logical_df['a'] & logical_df['b']
logical_df
```

```python
logical_df['OR'] = logical_df['a'] | logical_df['b']
logical_df
```

#### Conditionals:

```python
x = 0

if x > 0:
  print("Condition met: x is greater than 0")
else:
  print("The condition is not met")
```

#### Integration exercise:
>>#### Write a program that reads the names of the studies and assigns them to a modality:
>>>* X-ray, US, or CT

```python
## List of radiological studies:

radiology_studies = ['hand x-ray', 'chest x-ray',
                     'pelvis x-ray', 'abdominal ultrasound',
                     'soft tissue ultrasound', 'cervical ultrasound',
                     'chest computed tomography', 'abdominal computed tomography']
```

```python
for study in radiology_studies:
  # print(study)
  if "x-ray" in study:
    print(f"{study} : is an x-ray")
  elif "ultrasound" in study:
    print(f"{study} : is an ultrasound")
  else:
    print(f"{study} : is a computed tomography study")
```

#### Functions

* A function is a reusable and organized block of code that performs a specific task.

* Structure:
>>> function(parameters)

* Convert modality evaluation into a function:

1. Receives a variable.
2. Verifies that the variable is text (string).
3. Evaluates whether x-ray is found in the content of the entered word.
4. If not found, evaluates whether ultrasound is found in the content of the entered word.
5. If not found, assigns it to computed tomography.
6. Returns the final modality.

```python
## Syntax: def function(parameters)

def evaluate_modality(text):
  """
  Function that receives a text string and evaluates the
  presence of keywords related to the
  diagnostic modality.
  """
  if type(text) == str:
    if "x-ray" in text:
      modality = 'x-ray'
    elif "ultrasound" in text:
      modality = 'ultrasound'
    else:
       modality = 'computed tomography'
  else:
    print("The entered variable is not a string")

  return modality
```

```python
for study in radiology_studies:
  print(f"{study} corresponds to the modality:", evaluate_modality(study))
```
